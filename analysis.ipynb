#Step 3: Clean data to remove duplication and empty row




```python
#code clean data

import pandas as pd

# Load the CSV file into a Pandas DataFrame
Lazada = 'lazada.csv'
df_original = pd.read_csv(Lazada)

# Display the shape (rows, columns) of the original data
original_shape = df_original.shape

# Drop duplicates and rows with missing values
df_cleaned = df_original.drop_duplicates()
df_cleaned.dropna(inplace=True)

# Calculate the number of rows dropped
rows_dropped = original_shape[0] - df_cleaned.shape[0]

# Save the cleaned data to a new CSV file
cleaned_file_path = 'cleaned_file_1st attempt.csv'
df_cleaned.to_csv(cleaned_file_path, index=False)

# Display the difference in row count and the first few rows of both datasets
print("Original Data ({} rows):".format(original_shape[0]))
print(df_original.head())

print("\nCleaned Data ({} rows):".format(df_cleaned.shape[0]))
print(df_cleaned.head())

print("\n{} rows have been dropped during cleaning.".format(rows_dropped))
print("Cleaned data saved to: {}".format(cleaned_file_path))
```

# Step 6: Convert text to numeric data

```python
#Remove text symbol in data set final
import pandas as pd
import re

# Load the cleaned CSV data into a pandas DataFrame
data = pd.read_csv('cleaned_file_1st attempt.csv')

# Clean 'Price' column: remove non-numeric characters and convert to numeric
data['Price'] = data['Price'].str.replace(r'[^\d.]', '', regex=True).str.replace('.', '').astype(int)

# Clean 'Sold Quantity' column: remove 'Đã bán', commas, and replace 9999+ with 10000
data['Sold'] = data['Sold'].str.replace('Đã bán', '').str.replace(',', '').str.replace('+', '').fillna(0).astype(int)
data['Sold'] = data['Sold'].replace(9999, 10000)

# Clean 'Review Count' column: remove brackets and retain the number in between
data['Review'] = data['Review'].str.replace(r'[\(\)]', '', regex=True).astype(int)

# Print the cleaned DataFrame
print(data.head())

# Save the cleaned DataFrame to a new CSV file
data.to_csv('cleaned_file_2nd attempt.csv', index=False)
```

# Step 8: Plot chart by price & sold & review

```python
#to plot chart 33
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Remove leading/trailing whitespaces and special characters from column names
data.columns = data.columns.str.strip()  # Remove leading/trailing whitespaces
data.columns = data.columns.str.replace('[^a-zA-Z0-9]', '_', regex=True)  # Replace special characters with underscores

# Create histograms and box plots for numeric variables (price, sold, review)
numeric_columns = ['Price', 'Sold', 'Review']
for col in numeric_columns:
    plt.figure(figsize=(12, 6))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(data=data, x=col, bins=20, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    # Box plot
    plt.subplot(1, 2, 2)
    sns.boxplot(data=data, y=col)
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)

    plt.tight_layout()
    plt.show()

# Create a bar chart for the 'category' variable
plt.figure(figsize=(12, 6))
sns.countplot(data=data, y='Category', order=data['Category'].value_counts().index)
plt.title('Bar Chart of Category')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()

# Display a message indicating the completion of the analysis
print("Graphs generated successfully.")
```

# Step 9: Correlation between gift and sold


```python
#correlation between gift and sold

import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV data into a pandas DataFrame
data = pd.read_csv('cleaned_file_final.csv')

# Define a function to convert "Gift" to binary indicating presence of a gift
def convert_to_binary(gift):
    if gift == 'Quà tặng':
        return 1
    elif gift == 'Mua 1 giảm 40.000 đ':
        return 2
    else:
        return 0

# Convert "Gift" to binary indicating presence of a gift
data['Gift_binary'] = data['Gift'].apply(convert_to_binary)

# Calculate correlation between "Sold" and "Gift_binary"
correlation_sold_gift = data['Sold'].corr(data['Gift_binary'])

# Display the correlation
print("Correlation between Sold and Gift_binary:")
print(correlation_sold_gift)

# Plot a scatter graph to illustrate the correlation
plt.scatter(data['Gift_binary'], data['Sold'])
plt.xlabel('Gift_binary')
plt.ylabel('Sold')
plt.title('Scatter Plot of Sold vs Gift_binary')
plt.show()
```

# Step 9: Correlation between gift and sold and PRICE

```python
#Correlation between sold, gift and price của VK có 3 biên nè
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your CSV data into a pandas DataFrame
data = pd.read_csv('cleaned_file_final.csv')

# Define a function to convert "Gift" to binary indicating presence of a gift
def convert_to_binary(gift):
    if gift == 'Quà tặng':
        return 1
    elif gift == 'Mua 1 giảm 40.000 đ':
        return 22
    else:
        return 0

# Convert "Gift" to binary indicating presence of a gift
data['Gift_binary'] = data['Gift'].apply(convert_to_binary)

# Calculate correlation matrix for "Price", "Sold", and "Gift_binary"
correlation_matrix = data[['Price', 'Sold', 'Gift_binary']].corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Plot a scatter plot matrix
scatter_matrix = pd.plotting.scatter_matrix(data[['Price', 'Sold', 'Gift_binary']], figsize=(10, 8), diagonal='kde')

# Annotate the plots with variable names
for ax in scatter_matrix.ravel():
    ax.set_xlabel(ax.get_xlabel(), fontsize=12)
    ax.set_ylabel(ax.get_ylabel(), fontsize=12)

plt.suptitle('Scatter Plot Matrix', y=0.95, fontsize=16)
plt.show()


```
